```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
X = iris.data
y = iris.target
```

```python
iris_df = pd.DataFrame(X,columns=iris.feature_names)
iris_df['target'] = y
```

```python
num_names_dict = {idx: v for idx, v in enumerate(iris.target_names)}
num_names_dict
```

```python
iris_df['Species'] = iris_df['target'].map(num_names_dict)
```

```python
iris_df.head()
```

```python
iris_df['Species'].value_counts()
```

### 使用 DataFrame 提供的绘图功能

```python
iris_df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)')
plt.show()
```

## 查看两个变量的相关关系

```python
import seaborn as sns

# size 表示图片的显示大小 # pearsonr
sns.jointplot(x='sepal length (cm)',
              y='sepal width (cm)', data=iris_df, size=6)
plt.show()
```

## 绘制表格

```python
sns.FacetGrid(iris_df, hue='Species', size=6).map(plt.scatter, 'sepal length (cm)', 'sepal width (cm)').add_legend()
plt.show()
```

## 绘制箱线图（传入 DataFrame 对象）

```python
sns.boxplot(x='Species', y='sepal length (cm)', data=iris_df)
plt.show()
```

```python
sns.boxplot(x='Species', y='sepal width (cm)', data=iris_df)
plt.show()
```

## 绘制抖动图

```python
ax = sns.boxplot(x="Species", y="sepal width (cm)", data=iris_df)
ax = sns.stripplot(x="Species", y="sepal width (cm)", data=iris_df, jitter=True, edgecolor="gray")
plt.show()
```

## 小提琴图

```python
sns.violinplot(x="Species", y="sepal width (cm)", data=iris_df, size=5)
plt.show()
```

```python
iris_df.columns
```

```python
# 小提琴图结合了前面两个图的特点
# 看每个分类上，某个特征的分布情况
sns.violinplot(x="Species", y="petal length (cm)", data=iris_df, size=5)
plt.show()
```

## 绘制表格图

```python
# kernel density estimate 核密度估计，其实就是概率密度分布
sns.FacetGrid(iris_df, hue='Species', size=5).map(sns.kdeplot, 'petal length (cm)').add_legend()
plt.show()
```

## 成对变量显示

```python
%%time
sns.pairplot(iris_df.drop(['target'], axis=1), hue='Species')
plt.show()
```

## 更改对角线的图表类型

```python
%%time
sns.pairplot(iris_df.drop(['target'], axis=1), hue='Species', diag_kind='kde')
plt.show()
```

## 使用 pandas 绘制每个特征对于类别的分布

```python
iris_df.drop(['target'], axis=1).boxplot(by='Species', figsize=(12, 6))
plt.show()
```

```python
from pandas.tools.plotting import andrews_curves
andrews_curves(iris_df.drop("target", axis=1), "Species")
plt.show()
```

```python
from pandas.tools.plotting import parallel_coordinates

parallel_coordinates(iris_df.drop("target", axis=1), "Species")
```

```python
from pandas.tools.plotting import radviz

radviz(iris_df.drop("target", axis=1), "Species")
```
