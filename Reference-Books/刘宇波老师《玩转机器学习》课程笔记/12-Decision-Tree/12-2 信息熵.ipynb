# 12-2 信息熵

## 什么是信息熵？
+ 熵在信息论中代表随机变量的不确定性的度量
+ 熵越大，数据的不确定性越高
+ 熵越小，数据的不确定性越低

### 信息熵的计算公式

$$
H = -\sum_{i=1}^k p_i \cdot {\rm log}(p_i)
$$

### 例：计算 3 个例子来理解信息熵作为数据的不确定性的度量

说明：下面的计算中的 ${\rm log}$ 函数暂时以 $e$ 为底，一般来说，都以 2 为 底，单位是 bit。

```python
import math


def entropy(probabilities):
    # 如果概率是 0 ，就不参与计算了
    probabilities = filter(lambda x: x != 0, probabilities)
    result = 0
    for probability in probabilities:
        result += (-probability * math.log(probability))
    return result


probabilities = [1. / 10, 2. / 10, 7. / 10]
entropy1 = entropy(probabilities)

probabilities = [1.0 / 3, 1.0 / 3, 1.0 / 3]
entropy2 = entropy(probabilities)

probabilities = [1.0, 0, 0]
entropy3 = entropy(probabilities)

print(entropy1)
print(entropy2)
print(entropy3)
```

## 以二分类问题为例，画出信息熵函数的图像

$$
H = -x \cdot {\rm log}(x) - (1 - x) \cdot {\rm log}(1-x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy_2_class(p):
    return -p * np.log(p) - (1 - p) * np.log(1 - p)
```

```python
x = np.linspace(0.01, 0.99, 200)
plt.figure(dpi=100)
plt.plot(
    x,
    entropy_2_class(x),
    label='$H = -x \cdot log(x) - (1 - x) \cdot log(1-x)$')
plt.legend()
plt.show()
```

划分以后信息熵降低。
