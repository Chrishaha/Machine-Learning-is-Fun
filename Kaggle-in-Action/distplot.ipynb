```python
# IPython notebook 中的魔法方法，这样每次运行后可以直接得到图像，不再需要使用 plt.show()
%matplotlib inline 
# 导入numpy包，用于生成数组
import numpy as np  
import seaborn as sns  # 习惯上简写成 sns sns.set()           
sns.set()# 切换到 seaborn 的默认运行配置
```

```python
x = np.random.randint(1, 101, (100))  # 随机生成100个符合正态分布的数 sns.kdeplot(x)
```

```python
min(x), max(x)
```

```python
# cut：参数表示绘制的时候，切除带宽往数轴极限数值的多少，默认为 3
sns.kdeplot(x, cut=0, cumulative=True)
```

```python
sns.distplot(x, color="g")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3)  # 创建一个一行三列的画布
sns.distplot(x, ax=axes[0])  # 左图
sns.distplot(x, hist=False, ax=axes[1])  # 中图
sns.distplot(x, kde=False, ax=axes[2])  # 右图
```

```python
fig, axes = plt.subplots(1, 2)
sns.distplot(x, rug=True, ax=axes[0])  # 左图
sns.distplot(x, ax=axes[1])  # 右图
```

fit：控制拟合的参数分布图形，能够直观地评估它与观察数据的对应关系（黑色线条为确定的分布）。

```python
# norm_hist：若为 True , 则直方图高度显示密度而非计数，含有 kde 图像中默认为 True 

fig, axes = plt.subplots(1, 2)
sns.distplot(x, norm_hist=True, kde=False, ax=axes[0])  # 左图
sns.distplot(x, kde=False, ax=axes[1])  # 右图
```
