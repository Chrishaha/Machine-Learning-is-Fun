# 鸢尾花数据集可视化

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

```python
plt.scatter(X[y == 0, 2], X[y == 0, 3], c='r')
plt.scatter(X[y == 1, 2], X[y == 1, 3], c='b')
```
