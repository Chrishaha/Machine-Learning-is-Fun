# NumPy 的 meshgrid 函数的作用

meshgrid 的作用是根据传入的两个一维数组参数生成两个数组元素的列表。

如果第一个参数是 xarray，维度是 xdimesion ，第二个参数是 yarray ，维度是 ydimesion 。
那么生成的第一个二维数组是以 xarray 为行，ydimesion 行的向量；而第二个二维数组是以 yarray 的转置为列，xdimesion 列的向量。

```python
import numpy as np

xnums = np.arange(1, 3)
ynums = np.arange(10, 14)
```

```python
xnums
```

```python
ynums
```

```python
xx, yy = np.meshgrid(xnums, ynums)
```

## 把图画出来就很清晰了

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

m, n = (5, 3)
x = np.linspace(0, 4, m)
y = np.linspace(0, 2, n)

X, Y = np.meshgrid(x, y)
print('X', X.shape)
print('Y', Y.shape)

# plot 里面传矩阵也是可以的
plt.plot(X, Y, color='blue', marker='.', linestyle='none')
plt.show()
```

然后我们再获得数据点，看看图是怎么画出来的。

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 4)
y = np.arange(10, 13)

X, Y = np.meshgrid(x, y)

z = [i for i in zip(X.flat, Y.flat)]
for item in z:
    print(item)
```

这下你就很清楚，这个函数是在干嘛了吧，对滴，就是在画表格。从左下角，到右上角（先从左到右，在往上一行，再从左到右画表格）。
