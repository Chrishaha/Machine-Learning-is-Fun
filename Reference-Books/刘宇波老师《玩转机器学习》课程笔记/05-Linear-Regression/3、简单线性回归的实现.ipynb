# 实现简单的线性回归

## 本节要点：
+ 什么是简单线性回归？自变量只有一个未知数的回归问题，就是简单线性回归。
+ 如何求解简单线性回归？
+ 试着按照 scikit-learn 的实现模式自己实现以下简单线性回归。
+ 1、初始化；2、fit；3、predict。
+ 简单的线性回归是完全可以画出拟合曲线的。

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])

plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
# numerator 分子
# denominator 分母

num = 0.0
d = 0.0

for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean) ** 2

a = num / d
b = y_mean - a * x_mean
```

$a$ 是拟合曲线（回归曲线）的斜率。

```python
a
```

$b$ 是拟合曲线（回归曲线）的截距。

```python
b
```

下面我们画出这条拟合曲线。

```python
y_hat = a * x + b

plt.scatter(x, y)
plt.plot(x, y_hat, color = 'r')
plt.axis([0, 6, 0, 6])
plt.show()
```
