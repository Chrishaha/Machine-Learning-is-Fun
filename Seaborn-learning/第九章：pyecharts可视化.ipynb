```python
import numpy as np
import pandas as pd
# 可能需要安装 gcc，命令如下： conda install gcc
import pyecharts 
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [10, 20, 30, 40, 50, 60]
scatter = pyecharts.Scatter("散点图示例")
scatter.add('成绩散点图', x, y)
scatter
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [10, 20, 30, 40, 50, 60]
scatter = pyecharts.Scatter("散点图示例")
scatter.add(
    'A',
    x,
    y,
    is_visualmap=True,
    visual_type='size',
    visual_range_size=[10, 60])
scatter
```

```python
from pyecharts import Line

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [5, 20, 36, 10, 10, 100]
v2 = [55, 60, 16, 20, 15, 80]
line = Line("折线图示例")
line.add("商家A", attr, v1, mark_point=["average"])
line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average"])
line.render()
line
```

```python
years = ["1950", "1960", "1970", "1980", "1990", "2000", "2010"]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
line = pyecharts.Line("折线图示例")
line.add("GDP", years, gdp, mark_point=["average"])
line
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
years = [str(year) for year in years]

gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
line = pyecharts.Line("阶梯图")
line.add("GDP", years, gdp, is_step=True)
line
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
line = pyecharts.Line("面积图")
line.add("GDP", years, gdp, is_fill=True, area_color='#000', area_opacity=0.3)
line
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = {
    'sepal_length_cm': iris.data[:, 0],
    'sepal_width_cm': iris.data[:, 1],
    'petal_length_cm': iris.data[:, 2],
    'petal_width_cm': iris.data[:, 3]
}
iris_data = pd.DataFrame(data=data)
iris_data.head()
```

```python
# iris_data = pd.read_csv(open('./iris-data.csv'))
iris_data.head()
```

```python
x = list(iris_data.columns[0:4])
y = [
    list(iris_data.sepal_length_cm),
    list(iris_data.sepal_width_cm),
    list(iris_data.petal_length_cm),
    list(iris_data.petal_width_cm)
]

boxplot = pyecharts.Boxplot("箱形图")
y_data = boxplot.prepare_data(y)
boxplot.add('', x, y_data)
boxplot
```

```python
data = [
    ('上海', 78), ('武汉', 56), ('长沙', 45), ('北京', 65), ('苏州', 32), ('盐城', 15),
    ('南昌', 87), ('青岛', 45), ('广州', 78), ('拉萨', 12), ('桂林', 21), ('西安', 42),
    ('济南', 12)]

geo = pyecharts.Geo("地图绘制案例一",
                    title_pos="center", width=1200,
                    height=600)
attr, value = geo.cast(data)
geo.add("", attr, value, visual_range=[0, 100], visual_text_color="#fff",
        geo_normal_color="#FFFFFF",
        symbol_size=15, is_visualmap=True)
geo.render()
```

```python
data = [
    ('上海', 78), ('武汉', 56), ('长沙', 45), ('北京', 65), ('苏州', 32), ('盐城', 15),
    ('南昌', 87), ('青岛', 45), ('广州', 78), ('拉萨', 12), ('桂林', 21), ('西安', 42),
    ('济南', 12)]

geo = pyecharts.Geo("地图绘制案例二",
                    title_pos="center", width=1200,
                    height=600)
attr, value = geo.cast(data)
geo.add("", attr, value, type='heatmap', visual_range=[0, 100], visual_text_color="#fff",
        geo_normal_color="#FFFFFF",
        symbol_size=15, is_visualmap=True)
geo.render()
```
