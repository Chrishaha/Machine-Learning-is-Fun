 # DataFrame 在 contact 之后，一定要记得用 reset_index 去处理 index , 不然容易出现莫名的逻辑错误

或者在 `concat` 的时候，指定 `ignore_index = True`。

参考资料：https://blog.csdn.net/lujiandong1/article/details/52929090

```python
import pandas as pd

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})

df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']
})

frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

可以看到，直接拼在一起，索引会重复。
