# seaborn可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

+ 绘制散点图

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.scatter(years, gdp)
```

+ 设置 seaborn 的主题

```python
sns.set_style("darkgrid")
plt.scatter(years, gdp)
```

```python
# dark 这个主题就不会显示网格了
sns.set_style("dark")
plt.scatter(years, gdp)
```

### 更普遍地是使用 set 方法，可以设置主题、调色板、颜色代码 color_codes

```python
sns.set(style="white", palette="muted", color_codes=True)  # set( )设置主题，调色板，颜色主题更常用
plt.plot(np.arange(10))
plt.show()
```

## 坐标轴移除

+ 使用 sns.despine()，去掉了顶部和右侧。

```python
sns.set(style="white", palette="muted", color_codes=True)  # set( )设置主题，调色板更常用
plt.plot(np.arange(10))
sns.despine() # 把边框去掉了一些
```

设置偏移（好像没有什么用）。

```python
sns.set(style="white", palette="muted", color_codes=True)  # set( )设置主题，调色板更常用
plt.plot(np.arange(10))
sns.despine(offset=10, trim=True) # 设置边框偏移
```

```python
sns.set(style="whitegrid", palette="muted",
        color_codes=True)

# sns.set()
plt.plot(np.arange(10))
sns.despine(left=True, bottom=True)
```

# 单变量分布图

可以选择画直方图、概率密度曲线图还有毛坦图。

```python
tips = sns.load_dataset('tips')
tips.head()
```

数值型、连续型变量可以画单变量分布图。

```python
tips['total_bill'].dtype
```

```python
sns.set(color_codes=True)
sns.distplot(tips['total_bill']) # tips['total_bill'] 表示只有一个变量的数据
```

```python
# rug 表示显示毛毯图
sns.distplot(tips['total_bill'], kde=False, rug=True)
```

```python
sns.distplot(tips['total_bill'], bins=30, kde=False, rug=True)
```

```python
sns.distplot(tips['total_bill'], hist=False, rug=True)
```

## 学习 subplots 的用法（重要）

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.distplot(tips['total_bill'], ax=axes[0], kde=False)
sns.rugplot(tips['total_bill'], ax=axes[0])

sns.kdeplot(tips['total_bill'], ax=axes[1], shade=True)   # shade  阴影
sns.rugplot(tips['total_bill'], ax=axes[1])
```

# 多变量分布图

## 针对两个连续型、数值型变量，可以观察他们的相关关系

+ 使用 joinplot 还可以看他们的分布情况。

```python
sns.jointplot(x="tip", y="total_bill", data=tips)
```

+ 使用 kind='kde' ，默认的直方图就变成了密度图。

```python
# 图上还显示了相关系数
sns.jointplot(x="tip", y="total_bill", data=tips, kind='kde')
```

## 绘制成对的二元连续型变量分布图

```python
# 以轴矩阵的方式显示
# 小费数据多变量分布图
sns.pairplot(tips, size=2.5)
```

# 对于类别变量的数据展示

## 分类散点图

stripplot、swarmplot 可以看清每个数据点。

### 用一个类别变量去划分连续型变量

```python
sns.set(style="white", color_codes=True)  # 设置样式

sns.stripplot(x="day", y="total_bill", data=tips)
sns.despine()  # 去坐标轴
```

```python
# jitter=True 表示会有一些抖动，这样数值相近的点就不会挤在一起了
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, size=10)
sns.despine()
```

## swarmplot 比上面的 stripplot 加上抖动看得还更清楚一点

```python
sns.swarmplot(x="day", y="total_bill", data=tips, size=10)
sns.despine()
```

## hue="sex" 表示，在第 1 分类变量 day 的基础上，设置第 2 分类变量，即嵌套分类变量

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, size=10)
sns.despine()
```

## 箱线图可以观察每个分类下连续型变量的四分位数、中位数和极值

```python
# 箱线图，hue="time" 第 2 分类变量
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
sns.despine()
```

## 小提琴图是能让你看到分布的，把直方图给显示了出来

### hue="time" 第 2 分类变量，x 和 y 那个是分类变量，就决定了小提琴图的横向与纵向

```python
sns.set(style="whitegrid", color_codes=True)

sns.violinplot(x="total_bill", y="day", hue="time", data=tips)
```

# 这里 subplots 的用法上面也用到过

```python
sns.set(style="whitegrid", color_codes=True)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True, ax=axes[0])
# 对数据进行切分，inner="stick" 
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True, inner="stick", palette="Set3", ax=axes[1])
```

## 把之前的图进行组合，产生更强大的效果（相当于几种图叠加在一起显示了出来）

```python
sns.set(style="whitegrid", color_codes=True)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# 设置样式与子图

sns.violinplot(x="day", y="total_bill", data=tips,inner=None, ax=axes[0])
# inner内部不填充
sns.swarmplot(x="day", y="total_bill", data=tips, color="w", alpha=.5, ax=axes[0])

sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[1])
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, color="w", alpha=.5, ax=axes[1])
```

# 柱状图

## 这里的柱状图是使用了汇总的，默认的汇总方式是“mean”

```python
# 这里 x 是离散型变量
# hue 是第二分类变量

sns.barplot(x="sex", y="tip", hue="day", data=tips)
```

```python
# 更改汇总方式
sns.barplot(x="sex", y="tip", hue="day", data=tips, estimator=sum)
```

## 如果绘制柱状图，你的汇总方式是 count 计数，你就直接使用 countplot 好了

```python
tips.groupby("size").count()
```

```python
sns.countplot(x="size", data=tips, palette="Greens_d")
```

## 绘制回归图 regplot

```python
sns.set(style="darkgrid", color_codes=True)
sns.regplot(x="total_bill", y="tip", data=tips)
```

## lmplot

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
# 关闭置信区间
sns.lmplot(x="total_bill", y="tip", data=tips, ci=None)
```

## 还可以再传入一个分类变量 hue 

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

## lmplot 功能就很强大了，col="time" 表示列方向上设置一个类别变量

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips)
```

## col="time", row="sex" 表示在行列方向上都设置类别变量，所以就是你看到的 4 格表的样子

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips)
```

## 绘制网格

```python
tips['tip'].head()
```

```python
# 这里的 col 指定分类变量，取值是离散值
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip") # "tip" 是连续型变量，设置绘制直方图
```

## hue 指定图的内部分类

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()
```
