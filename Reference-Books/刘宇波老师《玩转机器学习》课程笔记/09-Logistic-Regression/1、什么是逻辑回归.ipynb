# 1、什么是逻辑回归

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## 认识 Sigmoid 函数

$$
\frac{1}{1+e^{-t}}
$$

```python
def sigmoid(t):
    return 1. / (1. + np.exp(-t))
```

```python
x = np.linspace(-10, 10, 500)

plt.figure(figsize=(8, 6), dpi=100)
plt.grid(b=True)
plt.plot(x, sigmoid(x), label='$(1+e^{-t})^{-1}$')
plt.legend()
plt.show()
```

我们看一看，这个函数是不是很符合我们上面所说的函数：

+ 区分类别的自变量的值是 $0$ ，概率边界值是 $0.5$。
+ 越趋近于正无穷，概率值越大，且趋近于 $1$。
+ 越趋近于负无穷，概率值越小，且趋近于 $0$。

### 总结

逻辑回归模型就是这样一个模型：
$$
{\rm sigmoid}(x) = \cfrac{1}{1 + e^{-\theta \cdot x}}
$$

逻辑回归 = 预测概率 + 分类。

Sigmoid 函数用于预测概率，分类按照如下规则做：人为定义一个概率阈值，大于等于这个阈值归为一类，否则归为另外一类。
