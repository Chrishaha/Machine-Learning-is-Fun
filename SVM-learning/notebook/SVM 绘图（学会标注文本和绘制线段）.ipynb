# SVM 绘图（学会标注文本和绘制线段）

```python
# SVM 绘图（学会标注文本和绘制线段）

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Support Vector Machine
class1 = np.array([[1, 1], [1, 3], [2, 1], [1, 2], [2, 2]])
class2 = np.array([[4, 4], [5, 5], [5, 4], [5, 3], [4, 5], [6, 4]])
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(class1[:, 0], class1[:, 1], c='r', marker='o')
plt.scatter(class2[:, 0], class2[:, 1], c='g', marker='s')
# 绘制决策边界
plt.plot([0, 6], [6, 0], c='r')

plt.plot([0, 4], [4, 0], color='r', linestyle='--')
plt.plot([1, 7], [7, 1], color='r', linestyle='--')
# x, y, dx, dy,
# 画垂直于 SVM 最佳分离超平面的线段
plt.arrow(4, 4, -1, -1, shape='full', color='r', linestyle='--')
# 画一个感知机算法得到的决策边界
plt.plot([3, 3], [0.5, 5], linestyle='--', color='b')
# 画垂直于一个感知机分类超平面的线段
plt.arrow(4, 4, -1, 0, shape='full', color='b', linestyle='--')

# 添加标注
# xy 箭头的坐标
# xytext 表示文字开始的坐标
# 标注 margin1
plt.annotate(r'margin 1',
             xy=(3.5, 4), xycoords='data',
             xytext=(3.1, 4.5), fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# 标注 margin2
plt.annotate(r'margin 2',
             xy=(3.5, 3.5), xycoords='data',
             xytext=(4, 3.5), fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# 标注两个支持向量
plt.annotate(r'support vector',
             xy=(4, 4), xycoords='data',
             xytext=(5, 4.5), fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.annotate(r'support vector',
             xy=(2, 2), xycoords='data',
             xytext=(0.5, 1.5), fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.xlim((0, 7))
plt.ylim((0, 7))
plt.show()
```
