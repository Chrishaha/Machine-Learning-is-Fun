# 5-1 简单线性回归

+ 思想简单，实现容易

损失函数、效用函数。

## 一类机器学习算法的基本思路

近乎所有的**参数学习算法**都是下面的套路。

找到参数 $a$ 和 $b$，使得目标函数尽可能小。

+ 通过分析问题，确定问题的损失函数或者效用函数。
+ 通过最优化损失函数或者效用函数，获得机器学习的模型。


## 简单线性回归的解决方法

**简单线性回归是典型的最小二乘法问题：通过求解最小化误差的平方，得到参数 $a$ 和 $b$ 关于训练数据集（特征和标签）的数学表达式。**


## 1 简单线性回归（样本特征只有一个的情况）

+ 解决回归问题
+ 思想简单，实现容易
+ **线性回归算法是许多强大的非线性模型的基础。**
+ **结果具有很好的可解释性**
+ 蕴含机器学习中的很多重要思想


什么是线性回归算法？样本特征和输出标记之间存在**线性关系**。
关键词：拟合
拟合就是寻找一条直线，最大程度“拟合”样本特征和输出标记之间的关系。

补充说明：对比我们上一章学习的分类问题。
+ 在分类的场景下，我们 plot 出图形的时候， $x$ 轴和 $y$ 轴都表示特征。
+ 在回归的场景下，$x$ 轴表示特征，$y$ 轴就表示目标值。特别说明，这是针对样本的特征数只有一个的时候，这种情况叫做“简单线性回归”。
**这是因为，在回归的问题中，我们要预测的数值是在一个连续的空间中，并不能像分类问题那样，用简单的几个标签就能够区分出来，所以回归问题的输出标签要占用一个坐标轴来表示。**


一个表达预测值与真实值差距的表达式：$(y^{(i)} - \hat y^{(i)})^2$.
+ 起到了绝对值的作用
+ 连续并且可导
目标：找到 $a$ 和 $b$ ，使得表达式最小。

解决这个问题，我们可以得到 $a$ 和 $b$ 的数学表达式（可以自己推导一遍）。

**所有的参数学习算法都是下面的这个套路。**
$(y^{(i)} - \hat y^{(i)})^2$ 这个函数称为损失函数。效用函数（utility function）度量的是拟合的程度。
通过分析问题，确定问题的损失函数或者效用函数，通过最优化损失函数或者效用函数，获得机器学习的模型。

与 kNN 算法的一个巨大的差别：线性回归算法是一个“参数学习算法”。
知识扩展：《最优化原理》、《凸优化》。
