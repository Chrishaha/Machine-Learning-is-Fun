# 插值函数的使用

```python
import numpy as np

x = np.linspace(0, 4, 12)
y = np.cos(x**2 / 3 + 4)
plt.plot(x, y, 'o')
plt.show()
```

## 引入 scipy 提供的 interpolate 模块

```python
from scipy import interpolate
```

```python
from scipy.interpolate import interp1d

f1 = interp1d(x, y, kind = 'linear')
f2 = interp1d(x, y, kind = 'cubic')

xline = np.linspace(0, 4, 50)
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', xline, f1(xline), '-')
plt.show()
```

```python
from scipy.interpolate import interp1d

# 线性插值
f1 = interp1d(x, y, kind = 'linear')
# 三次样条插值
f2 = interp1d(x, y, kind = 'cubic')

xline = np.linspace(0, 4, 50)
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', xline, f2(xline), '--')
plt.show()
```
