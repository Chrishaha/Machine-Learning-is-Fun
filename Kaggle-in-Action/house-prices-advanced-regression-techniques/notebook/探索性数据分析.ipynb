# 探索性数据分析

```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('../input/train.csv', index_col=0)
df_test = pd.read_csv('../input/test.csv', index_col=0)
```

## 目标变量对数化使之符合正态分布

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 在我的 notebook 里，要设置下面两行才能显示中文
plt.rcParams['font.family'] = ['sans-serif']
# 如果是在 PyCharm 里，只要下面一行，上面的一行可以删除
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

### 使用 Pandas 频率分布直方图

```python
prices = pd.DataFrame({
    "price": df_train.SalePrice,
    "log(price + 1)": np.log1p(df_train.SalePrice)
})
prices.hist()
plt.show()
```

### 使用 Seaborn 绘制频率分布直方图

```python
fig = plt.figure(figsize=(20, 6))

plt.subplot(121)
plt.title('原始目标变量', fontdict={'size': 20})
sns.distplot(df_train.SalePrice)

plt.subplot(122)
plt.title('对数化以后的目标变量', fontdict={'size': 20})
sns.distplot(np.log1p(df_train.SalePrice))

plt.show()
```

> 符合正态分布的数据，在使用线性回归的时候，效果会更好。
