```python
y_train = train["label"]
X_train = train.drop(labels=["label"], axis=1)
```

```python
def show_pic(X, y):
    plt.figure(figsize=(12, 10))
    for i in range(25):
        plt.subplot(5, 5, i + 1)
        plt.imshow(X.loc[i].values.reshape(28, 28), cmap=plt.cm.gray)
        plt.title('label:{}'.format(y.loc[i]), fontsize=14)
        plt.xticks(())
        plt.yticks(())
    plt.show()


show_pic(X_train, y_train)
```

```python
def show_one_pic(idx):
    img = X_train.iloc[idx].values
    img = img.reshape((28, 28))
    plt.imshow(img, cmap='gray')
    plt.title('label = {}'.format(y_train[idx]), fontsize=20)
    plt.axis("off")
    plt.show()
```
