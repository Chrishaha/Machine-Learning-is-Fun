# 使用网格搜索寻找 $k$ 近邻算法最佳超参数

## 加载数据

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target
```

## 分割训练集与测试数据集

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=666)
```

## 超参数以列表形式给出，列表里面的单个元素是一个字典

```python
param_grid = [
    {
        'weights': ['uniform'],
        'n_neighbors': [i for i in range(1, 11)]
    },
    {
        'weights': ['distance'],
        'n_neighbors': [i for i in range(1, 11)],
        'p': [i for i in range(1, 6)]

    }
]
```

```python
from sklearn.model_selection import GridSearchCV

# 看一看网格搜索这个类的用法
GridSearchCV?
```

## 在进行网格搜索的时候，GridSearchCV 这个类会帮我们做交叉验证

+ 可以通过设置 `cv` 这个参数设置交叉验证的折数

```python
%%time
from sklearn.neighbors import KNeighborsClassifier


knn_clf = KNeighborsClassifier()
# n_jobs = -1 将计算机所有的核都用于并行计算
# verbose 在搜索的过程中进行一些输出，整数值越大，输出的信息越详细，通常选择 2
grid_search = GridSearchCV(knn_clf, param_grid, cv=10, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
# estimator 评估量
# 网格搜索最佳分类器的超参数的值
print(grid_search.best_estimator_)
```

```python
knn_clf = grid_search.best_estimator_
knn_clf.predict(X_test)
knn_clf.score(X_test, y_test)
```
