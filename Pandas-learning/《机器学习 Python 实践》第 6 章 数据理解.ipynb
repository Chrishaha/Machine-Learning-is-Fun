```python
plas_demean = data['plas'] - data['plas'].mean()
age_demean = data['age'] - data['age'].mean()
```

```python
# 协方差
covariance = plas_demean.values.dot(age_demean.values)
```

```python
# 方差
std = np.linalg.norm(plas_demean.values) * np.linalg.norm(age_demean.values)
```

```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
correlations = data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
from numpy import set_printoptions
from sklearn.preprocessing import Binarizer

array = data.values
X = array[:, 0:8]
Y = array[:, 8]
transform = Binarizer(threshold=0.0).fit(X)
# 数据转换
newX = transform.transform(X)
# 设定数据的打印格式
set_printoptions(precision=3)
print(newX)
```
