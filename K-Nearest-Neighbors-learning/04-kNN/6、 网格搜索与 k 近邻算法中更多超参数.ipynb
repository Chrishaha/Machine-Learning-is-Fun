# 6、网格搜索与 k 近邻算法中更多超参数

网格搜索之前的准备。

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split


digits = datasets.load_digits()
X = digits.data
y = digits.target

# 分割训练数据集与测试数据集。
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=666)
```

**使用网格搜索之前，先实例化我们的 KNeighborsClassifier 对象（不带任何参数）。**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# 网格搜索的参数形式是列表，列表中元素是字典，字典的 key 是方法的参数
param_grid = [
    {
        'weights': ['uniform'],
        'n_neighbors':[i for i in range(1, 11)]
    },
    {
        'weights': ['distance'],
        'n_neighbors':[i for i in range(1, 11)],
        'p':[i for i in range(1, 6)]

    }
]

knn_clf = KNeighborsClassifier()
grid_search = GridSearchCV(knn_clf, param_grid, verbose=2)
```

```python
%%time
grid_search.fit(X_train, y_train)
```

### 三个重要的 grid_search 属性

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
# estimator 评估量
# 网格搜索最佳分类器的超参数的值
grid_search.best_estimator_
```

**下面，我们就可以用那个最好的分类器，去运行我们的数据。**

```python
knn_clf = grid_search.best_estimator_
knn_clf
```

```python
knn_clf.predict(X_test)
```

```python
knn_clf.score(X_test, y_test)
```

## GridSearchCV 其它参数的含义

```python
%%time

# n_jobs = -1 将计算机所有的核都用于并行计算
# verbose 在搜索的过程中进行一些输出，整数值越大，输出的信息越详细，通常选择 2
grid_search = GridSearchCV(knn_clf, param_grid, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, y_train)
```
