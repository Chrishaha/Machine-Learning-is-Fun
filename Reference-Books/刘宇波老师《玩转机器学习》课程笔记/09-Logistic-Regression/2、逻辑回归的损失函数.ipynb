# 2、逻辑回归的损失函数

```python
import numpy as np
import matplotlib.pyplot as plt

epsilon = 1e-06
x = np.linspace(0 + epsilon, 1 - epsilon, 100)
plt.figure(figsize=(10, 8))
plt.subplot(121)
plt.plot(x, -np.log(x), label='$y=-log(x)$')
plt.legend(loc='lower left', fontsize=20)
plt.xlim(-.5, 2)
plt.ylim(-.5, 2)
plt.title('当 y = 1 时，损失函数图像', fontsize=20)
plt.grid(b=True)

plt.subplot(122)
plt.plot(x, -np.log(1 - x), label='$y=-log(1-x)$')
plt.xlim(-.5, 2)
plt.ylim(-.5, 2)
plt.title('当 y = 0 时，损失函数图像', fontsize=20)
plt.grid(b=True)
plt.legend(loc='lower right', fontsize=20)
plt.show()
```

综上，我们的损失函数可以用一个数学表达式来表示，就是：
$$
{\rm cost}(x) = y \cdot [-{\rm log}(x)] + (1 - y) \cdot [-{\rm log}(1 - x)]
$$

那么，对于所有的样本而言，损失函数就应该定义为上述单个样本损失函数的期望，即：

$$
J(\theta) = \frac{1}{m}\sum_{i=1}^{m} \{ y^{(i)} \cdot [-{\rm log}(\hat p^{(i)})] + (1 - y^{(i)}) \cdot [-{\rm log}(1 - \hat p^{(i)})]\} 
$$

liwei 注：

1、这个损失函数其实是著名的交叉熵损失函数，用于度量分类问题的损失；

2、损失函数极小化等价于极大似然估计，即：可以使用统计学中的极大似然估计法得到逻辑回归的损失函数。

此时 $\theta$ 是自变量，是我们要训练的参数。我们经过分析得到的损失函数没有数学解析解，不过好在这个损失函数是一个凸函数，使用梯度下降法求得的局部最优解就是唯一的全局最优解。
